```python
import requests
import pandas as pd
```

```python
url='https://itunes.apple.com/search'
params = {'media':'music', 'term':'manizhe', 'limit':30}
response= requests.get(url, params=params)
response
```

```python
response.status_code
```

```python
response.text
```

```python
response.json()
```

```python
results = response.json()['results']
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.to_excel(r'C:\Users\user\Untitled Folder 1\data\iTunes_api.xlsx', sheet_name = 'manizha', index_label= 'id', encoding='utf-8')
```
